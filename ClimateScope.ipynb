Identify missing values, anomalies, and data coverage.

```python
import numpy as np


# ----------------------------
# 1. Missing Values
# ----------------------------
print("Missing values per column:")
print(df.isnull().sum())
print("\n% of missing values:")
print((df.isnull().sum() / len(df)) * 100)

# ----------------------------
# 2. Anomalies (basic checks)
# ----------------------------

# Example: Temperature outside realistic range (-90°C to 60°C)
if "temperature" in df.columns:
    anomalies_temp = df[(df["temperature"] < -90) | (df["temperature"] > 60)]
    print(f"\n Temperature anomalies: {len(anomalies_temp)} records")

# Example: Humidity should be between 0–100%
if "humidity" in df.columns:
    anomalies_humidity = df[(df["humidity"] < 0) | (df["humidity"] > 100)]
    print(f"\n Humidity anomalies: {len(anomalies_humidity)} records")

# Example: Wind speed negative?
if "wind_speed" in df.columns:
    anomalies_wind = df[df["wind_speed"] < 0]
    print(f"\n Negative wind speed values: {len(anomalies_wind)} records")

# ----------------------------
# 3. Data Coverage
# ----------------------------
print("\ Dataset Coverage:")
print("Time Range (if 'date' column exists):")
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
    print("Start:", df["date"].min())
    print("End:", df["date"].max())

print("\nGeographical Coverage (if 'city' or 'country' exists):")
if "country" in df.columns:
    print("Unique Countries:", df["country"].nunique())
if "city" in df.columns:
    print("Unique Cities:", df["city"].nunique())

print("\n Data inspection complete.")
```

Handle missing or inconsistent entries.

```python
# ----- A. Drop columns with too many missing values (e.g. >40%) -----
threshold = 0.4
missing_fraction = df.isnull().mean()
cols_to_drop = missing_fraction[missing_fraction > threshold].index
df.drop(columns=cols_to_drop, inplace=True)
print(f"🗑️ Dropped columns: {list(cols_to_drop)}")

# ----- B. Fill missing numerical values with mean/median -----
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.median()))

# ----- C. Fill missing categorical values with mode -----
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))

```

Convert units and normalize values.

```python
# Standardize categorical text (e.g., country names)
if "country" in df.columns:
    df["country"] = df["country"].str.strip().str.title()

# Remove impossible humidity values (<0 or >100)
if "humidity" in df.columns:
    df.loc[(df["humidity"] < 0) | (df["humidity"] > 100), "humidity"] = None
    df["humidity"] = df["humidity"].fillna(df["humidity"].median())

# Remove impossible temperature values (<-90 or >60)
if "temperature" in df.columns:
    df.loc[(df["temperature"] < -90) | (df["temperature"] > 60), "temperature"] = None
    df["temperature"] = df["temperature"].fillna(df["temperature"].median())

# Remove negative wind speeds
if "wind_speed" in df.columns:
    df.loc[df["wind_speed"] < 0, "wind_speed"] = None
    df["wind_speed"] = df["wind_speed"].fillna(df["wind_speed"].median())

print(" Remaining missing values:")
print(df.isnull().sum())

print("\n Cleaned dataset preview:")
print(df.head())
```

```python
# Temperature: Already have Celsius, keep it
df["temperature"] = df["temperature_celsius"]

# Wind speed: convert kph → m/s (1 kph = 0.27778 m/s)
df["wind_speed_ms"] = df["wind_kph"] * 0.27778

# Wind gusts: kph → m/s
df["gust_ms"] = df["gust_kph"] * 0.27778

# Pressure: keep in millibars (mb ~ hPa), drop inches
df["pressure_hpa"] = df["pressure_mb"]

# Precipitation: keep mm
df["precipitation_mm"] = df["precip_mm"]

# Visibility: keep in km
df["visibility_km"] = df["visibility_km"]

# Drop redundant unit columns if not needed
drop_cols = [
    "temperature_fahrenheit", "wind_mph", "gust_mph",
    "pressure_in", "precip_in", "visibility_miles"
]
df.drop(columns=drop_cols, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

cols_to_normalize = [
    "temperature", "wind_speed_ms", "gust_ms",
    "pressure_hpa", "precipitation_mm",
    "humidity", "visibility_km", "uv_index"
]

# Keep only columns that exist
cols_to_normalize = [c for c in cols_to_normalize if c in df.columns]

df_normalized = df.copy()
df_normalized[cols_to_normalize] = scaler.fit_transform(df[cols_to_normalize])

# Preview normalized data
print(df_normalized[cols_to_normalize].head())
```

Aggregate or filter data (e.g., daily to monthly averages)

```python
# Ensure datetime format
df["last_updated"] = pd.to_datetime(df["last_updated"], errors="coerce")

# Extract date, month, year
df["year"] = df["last_updated"].dt.year
df["month"] = df["last_updated"].dt.month
df["day"] = df["last_updated"].dt.day
```

```python
# Choose numeric columns for aggregation
numeric_cols = ["temperature", "wind_speed_ms", "gust_ms",
                "pressure_hpa", "precipitation_mm",
                "humidity", "visibility_km", "uv_index"]

# Monthly averages (global)
monthly_avg = df.groupby(["year", "month"])[numeric_cols].mean().reset_index()
print(monthly_avg.head())
```

```python
# Monthly averages per country
monthly_country_avg = df.groupby(
    ["country", "year", "month"]
)[numeric_cols].mean().reset_index()

print(monthly_country_avg.head())
```

Milestone 2: Core Analysis & Visualization Design (Weeks 2-4)

Perform statistical analysis to understand distributions, correlations, seasonal
patterns, and trends.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
numeric_cols = ["temperature", "humidity", "wind_speed_ms", "pressure_hpa", "precipitation_mm"]

df[numeric_cols].hist(figsize=(12,8), bins=30)
plt.suptitle("Distributions of Weather Variables")
plt.show()

# Kernel Density Estimate (KDE) for temperature
sns.kdeplot(df["temperature"], shade=True)
plt.title("Temperature Distribution")
plt.show()
```

```python
# Correlation matrix
corr = df[numeric_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Monthly averages (already computed before)
monthly_avg = df.groupby(["year","month"])[numeric_cols].mean().reset_index()

# Plot temperature trend across months
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_avg, x="month", y="temperature", hue="year", marker="o")
plt.title("Seasonal Temperature Patterns (Monthly Averages)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Prepare monthly averages + rolling mean
df["last_updated"] = pd.to_datetime(df["last_updated"], errors="coerce")
monthly_avg = (df.groupby(df["last_updated"].dt.to_period("M"))["temperature"]
                 .mean().reset_index())
monthly_avg["last_updated"] = monthly_avg["last_updated"].dt.to_timestamp()
monthly_avg["trend"] = monthly_avg["temperature"].rolling(6).mean()

# Plot
plt.figure(figsize=(14,7))
sns.lineplot(x="last_updated", y="temperature", data=monthly_avg,
             label="Monthly Avg", marker="o", color="skyblue")
sns.lineplot(x="last_updated", y="trend", data=monthly_avg,
             label="6-Month Trend", color="red", linewidth=2)

plt.axhline(monthly_avg["temperature"].mean(), ls="--", c="green",
            label=f"Overall Avg: {monthly_avg['temperature'].mean():.2f}°C")

plt.title("🌍 Global Average Temperature Trend", fontsize=16, weight="bold")
plt.xlabel("Year"); plt.ylabel("Temperature (°C)")
plt.grid(True, ls="--", alpha=0.5); plt.legend()
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.show()
```
